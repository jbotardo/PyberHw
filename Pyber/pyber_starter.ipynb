```python
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# File to Load (Remember to change these)
city_data_to_load = "data/city_data.csv"
ride_data_to_load = "data/ride_data.csv"

# Read the City and Ride Data
city_df = pd.read_csv(city_data_to_load)
ride_df = pd.read_csv(ride_data_to_load)

# Combine the data into a single dataset
city_df = city_df.drop_duplicates('city', keep ='first')
merged_df = city_df.merge(ride_df, on = 'city')
# Display the data table for preview
merged_df.head()
```

## Bubble Plot of Ride Sharing Data

```python
# Obtain the x and y coordinates for each of the three city types
urban_df = merged_df[merged_df['type'] == 'Urban'].groupby([merged_df["city"]])
suburban_df = merged_df[merged_df['type'] == 'Suburban'].groupby([merged_df["city"]])
rural_df = merged_df[merged_df['type'] == 'Rural'].groupby([merged_df["city"]])

urban_fare = urban_df["fare"].mean()
urban_ride_count = urban_df["ride_id"].count()
urban_driver_count = urban_df["driver_count"].mean()

suburban_fare = suburban_df["fare"].mean()
suburban_ride_count = suburban_df["ride_id"].count()
suburban_driver_count = suburban_df["driver_count"].mean()

rural_fare = rural_df["fare"].mean()
rural_ride_count = rural_df["ride_id"].count()
rural_driver_count = rural_df["driver_count"].mean()


# Build the scatter plots for each city types
plt.scatter(urban_ride_count, urban_fare, s=urban_driver_count * 10, color = 'r', edgecolor = 'black', label = 'Urban', alpha = 0.75)
plt.scatter(suburban_ride_count, suburban_fare, s= suburban_driver_count * 10, color = 'b', edgecolor = 'black', label = 'Suburban', alpha = 0.75)
plt.scatter(rural_ride_count, rural_fare, s= rural_driver_count * 10, color = 'y', edgecolor = 'black', label = 'Rural', alpha = 0.75)


# Incorporate the other graph properties
plt.title("Pyber Ride Sharing Data(2016)")
plt.xlabel("Total Number of Rides (Per City)")
plt.ylabel("Average Fare ($)")
# Create a legend
legend = plt.legend(title= "City Types")

# Incorporate a text label regarding circle size
plt.savefig("../Images/Pyber.png")
# Save Figure
plt.show()
```

```python
# Show plot
plt.show()
```

## Total Fares by City Type

```python
# Calculate Type Percents
grouped_type = merged_df.groupby(['type'])
total_fare = grouped_type['fare'].sum()

# Build Pie Chart
labels = ["Rural", "Suburban", "Urban"]
colors = ["gold","skyblue","salmon"]
explode = (0,0,0.1)
plt.pie(total_fare, explode=explode, labels=labels,colors=colors,autopct="%1.1f%%",shadow=True,startangle=150)
plt.title("% of Total Fares By City Type")
# Save Figure
plt.savefig("../Images/PyberPie.png")
```

```python
# Show Figure
plt.show()
```

## Total Rides by City Type

```python
# Calculate Ride Percents
rides_count = grouped_type['ride_id'].count()

# Build Pie Chart
labels = ["Rural", "Suburban", "Urban"]
colors = ["gold","skyblue","salmon"]
explode = (0,0,0.1)
plt.pie(rides_count, explode=explode, labels=labels,colors=colors,autopct="%1.1f%%",shadow=True,startangle=150)
plt.title("% of Total Fares By City Type")
# Save Figure
plt.savefig("../Images/PyberPie2.png")
```

```python
# Show Figure
plt.show()
```

## Total Drivers by City Type

```python
# Calculate Driver Percents
drivers_grouped = city_df.groupby(['type'])
total_drivers = drivers_grouped['driver_count'].sum()

# Build Pie Charts
labels = ["Rural", "Suburban", "Urban"]
colors = ["gold","skyblue","salmon"]
explode = (0,0,0.1)
plt.pie(total_drivers, explode=explode, labels=labels,colors=colors,autopct="%1.1f%%",shadow=True,startangle=150)
plt.title("% of Total Fares By City Type")

# Save Figure
plt.savefig("../Images/PyberPie3.png")
```
